creating function: function does not executed untill it is called.

```python
from tkinter import* #program starts here

root =Tk() #execute this line

def myClick(): #read it into memory. so later on, if you want to run it, you can call it and it will run
    myLabel = Label(root, text="Look! I clicked a Button!!")
    myLabel.pack()
    
    
myButton =Button(root, text="Click Me!", command =myClick, fg="blue", bg="yellow") #calling function by name:myClick
myButton.pack()

root.mainloop()

```

Creating Input Field:Entry widget

```python
from tkinter import*

root =Tk()

e=Entry(root, width=50,bg="black",fg="white",borderwidth=5)
e.pack()

def myClick(): 
    myLabel = Label(root, text="Hello" "  " + e.get()) #get function to get
    myLabel.pack()
    
    
myButton =Button(root, text="Enter Your Name", command =myClick, fg="blue", bg="yellow") 
myButton.pack()

root.mainloop()
```

OTHER WAY

```python
from tkinter import*

root =Tk()

e=Entry(root, width=50,bg="black",fg="white",borderwidth=5)
e.pack()
e.insert(0, "Enter Your Name:  " )

def myClick():
    hello= "Hello" "  " + e.get()
    myLabel = Label(root, text=hello) #using variable "hello"
    myLabel.pack()
    
    
myButton =Button(root, text="Enter Your Name", command =myClick, fg="blue", bg="yellow") 
myButton.pack()

root.mainloop()
```

```python
from tkinter import*

root =Tk()
root.title("Simple Calculator")

e=Entry(root, width=25,borderwidth=5,bg="yellow",fg="black")
e.grid(row=0,column=0,columnspan=3,padx=10,pady=10)

def button_click(number): #passing parameter through a function
    current =e.get()
    e.delete(0,END)
    e.insert(0, str(current) + str(number)) ##make sure pass string not the numbers because of +/concatenate
    
    
def button_clear():
    e.delete(0,END)
    
def button_add():
    first_number=e.get()
    global f_num          #f_num is global,we can use it outside of this function
    f_num=int(first_number)
    e.delete(0,END)

def button_equal():
    second_number=e.get() #get a number
    e.delete(0,END) #delete that number
    e.insert(0,f_num + int(second_number)) #then insert two numbers together
       

#Define buttons

button_1=Button(root,text="1",padx=40,pady=20,command=lambda: button_click(1))
button_2=Button(root,text="2",padx=40,pady=20,command=lambda: button_click(2))
button_3=Button(root,text="3",padx=40,pady=20,command=lambda: button_click(3))
button_4=Button(root,text="4",padx=40,pady=20,command=lambda: button_click(4))
button_5=Button(root,text="5",padx=40,pady=20,command=lambda: button_click(5))
button_6=Button(root,text="6",padx=40,pady=20,command=lambda: button_click(6))
button_7=Button(root,text="7",padx=40,pady=20,command=lambda: button_click(7))
button_8=Button(root,text="8",padx=40,pady=20,command=lambda: button_click(8))
button_9=Button(root,text="9",padx=40,pady=20,command=lambda: button_click(9))
button_0=Button(root,text="0",padx=40,pady=20,fg="blue",bg="skyblue",command=lambda: button_click(0))

button_add=Button(root,text="+",padx=39,pady=20,command=button_add)
button_equal=Button(root,text="=",padx=91,pady=20,command=button_equal)
button_clear=Button(root,text="CLEAR",padx=79,pady=20,fg="red",bg="lightgreen",command=button_clear)


#Put the button on the screen

button_1.grid(row=3,column=0)
button_2.grid(row=3,column=1)
button_3.grid(row=3,column=2)

button_4.grid(row=2,column=0)
button_5.grid(row=2,column=1)
button_6.grid(row=2,column=2)

button_7.grid(row=1,column=0)
button_8.grid(row=1,column=1)
button_9.grid(row=1,column=2)

button_0.grid(row=4,column=0)
button_clear.grid(row=4,column=1,columnspan=2)
button_add.grid(row=5,column=0)
button_equal.grid(row=5,column=1,columnspan=2)

    
    

root.mainloop()
```

How "=" button knows we want adding/multiplication/subtracting

```python
from tkinter import*

root =Tk()
root.title("Simple Calculator")

e=Entry(root, width=25,borderwidth=5,bg="yellow",fg="black")
e.grid(row=0,column=0,columnspan=3,padx=10,pady=10)

def button_click(number): #passing parameter through a function
    current =e.get()
    e.delete(0,END)
    e.insert(0, str(current) + str(number)) ##make sure pass string not the numbers because of +/concatenate
    
    
def button_clear():
    e.delete(0,END)
    
def button_add():
    first_number=e.get()
    global f_num         
    global math
    math ="addition"
    f_num=int(first_number)
    e.delete(0,END)

def button_equal():
    second_number=e.get()
    e.delete(0,END)
    
    if math=="addition":
             e.insert(0,f_num + int(second_number))
            
    if math=="subtraction":
             e.insert(0,f_num - int(second_number))
            
    if math=="multiplication":
             e.insert(0,f_num * int(second_number))
            
    if math=="division":
             e.insert(0,f_num / int(second_number))
            

    
def button_subtract():
    first_number=e.get()
    global f_num         
    global math
    math ="subtraction"
    f_num=int(first_number)
    e.delete(0,END)

def button_multiply():
    first_number=e.get()
    global f_num         
    global math
    math ="multiplication"
    f_num=int(first_number)
    e.delete(0,END)

def button_divide():
    first_number=e.get()
    global f_num         
    global math
    math ="division"
    f_num=int(first_number)
    e.delete(0,END)


       

#Define buttons

button_1=Button(root,text="1",padx=40,pady=20,command=lambda: button_click(1))
button_2=Button(root,text="2",padx=40,pady=20,command=lambda: button_click(2))
button_3=Button(root,text="3",padx=40,pady=20,command=lambda: button_click(3))
button_4=Button(root,text="4",padx=40,pady=20,command=lambda: button_click(4))
button_5=Button(root,text="5",padx=40,pady=20,command=lambda: button_click(5))
button_6=Button(root,text="6",padx=40,pady=20,command=lambda: button_click(6))
button_7=Button(root,text="7",padx=40,pady=20,command=lambda: button_click(7))
button_8=Button(root,text="8",padx=40,pady=20,command=lambda: button_click(8))
button_9=Button(root,text="9",padx=40,pady=20,command=lambda: button_click(9))
button_0=Button(root,text="0",padx=40,pady=20,fg="blue",bg="skyblue",command=lambda: button_click(0))

button_add=Button(root,text="+",padx=39,pady=20,command=button_add)
button_equal=Button(root,text="=",padx=91,pady=20,command=button_equal)
button_clear=Button(root,text="CLEAR",padx=79,pady=20,fg="red",bg="lightgreen",command=button_clear)

button_subtract=Button(root,text="-",padx=41,pady=20,command=button_subtract)
button_multiply=Button(root,text="*",padx=40,pady=20,command=button_multiply)
button_divide=Button(root,text="/",padx=41,pady=20,command=button_divide)



#Put the button on the screen

button_1.grid(row=3,column=0)
button_2.grid(row=3,column=1)
button_3.grid(row=3,column=2)

button_4.grid(row=2,column=0)
button_5.grid(row=2,column=1)
button_6.grid(row=2,column=2)

button_7.grid(row=1,column=0)
button_8.grid(row=1,column=1)
button_9.grid(row=1,column=2)

button_0.grid(row=4,column=0)
button_clear.grid(row=4,column=1,columnspan=2)
button_add.grid(row=5,column=0)
button_equal.grid(row=5,column=1,columnspan=2)

button_subtract.grid(row=6,column=0)
button_multiply.grid(row=6,column=1)
button_divide.grid(row=6,column=2)


    
    

root.mainloop()
```
